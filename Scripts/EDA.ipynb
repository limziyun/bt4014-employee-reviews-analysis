```python
amazon_df=data.loc[data['company']=='amazon']
apple_df=data.loc[data['company']=='apple']
facebook_df=data.loc[data['company']=='facebook']
google_df=data.loc[data['company']=='google']
microsoft_df=data.loc[data['company']=='microsoft']
netflix_df=data.loc[data['company']=='netflix']
```

### Average ratings per company

```python
all_columns = ['overall-ratings', 'work-balance-stars', 'culture-values-stars', 'carrer-opportunities-stars',
              'comp-benefit-stars', 'senior-mangemnet-stars']
all_df = [amazon_df, apple_df, facebook_df, google_df, microsoft_df, netflix_df, data]

df = pd.DataFrame()

for i in range(len(all_df)):
    for col in all_columns:
        tempdf = all_df[i][col]
        if (tempdf.dtype != 'float64'):
            tempdf = tempdf[tempdf != 'none']
            tempdf = pd.to_numeric(tempdf)
        x = tempdf.mean()
        if i==0:
            df.loc['Amazon', col] = x
        elif i==1:
            df.loc['Apple', col] = x
        elif i==2:
            df.loc['Facebook', col] = x
        elif i==3:
            df.loc['Google', col] = x
        elif i==4:
            df.loc['Microsoft', col] = x
        elif i==5:
            df.loc['Netflix', col] = x
        elif i==6:
            df.loc['All', col] = x
```

```python
df
```

```python
x = df.index.tolist()

for col in df:
    y = df[col]
    total = y[6]

    width = 1/1.5
    fig = plt.figure(figsize=(9, 6))
    ax = fig.add_subplot(111)
    ax.bar(x, y, color='c')
    ax.axhline(total, color="gray")
    ax.text(1.02, total, str(total), va='center', ha="left", bbox=dict(facecolor="w",alpha=0.5),
            transform=ax.get_yaxis_transform())
    ax.set_title(col)
    plt.show()
```
